```python
# Import all libraries needed
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt
from matplotlib import cm
```

# Caldeira-Leggett model
## Numerical integration of classical equations

### Preliminary test: harmonic oscillator

In order to test the correctness of VV implementation we first consider the case of a simple harmonic oscillator:

$$ \ddot{x}(t) + \omega^2x(t) = 0 \; ; $$

taking $x(0)=0$, $v(0)=\dot{x}(0)=1$ as initial conditions and choosing $\omega=1$ the exact solution reads:

$$ x(t) = \cos{t} \quad . $$

The particle's acceleration is given by the elastic force:

$$ F = -\omega^2 x \quad .$$

```python
# Implementation of Velocity-Verlet time integration scheme
    
# This function is written in such a way it can accept as input both scalars and vectors, so it would
# work for a system of particles too.

def velocity_verlet(x_old, v_old, acc, dt):
    a_old = acc(x_old)
    x_new = x_old + dt*v_old + 0.5*dt*dt*a_old
    a_new = acc(x_new)
    v_new = v_old + 0.5*dt*( a_old + a_new )
    return x_new, v_new
```

```python
# Numerical parameters and acceleration
omega = 1.0
acc = lambda x : -(omega*omega)*x

# Time step and time window
dt = 0.1
T = 25.0
N = int(T/dt)
t = np.linspace(0.0,T,N+1)

# Initial conditions and data structures initialization
x_0 = 0.0
v_0 = 1.0
x = np.empty(N+1)
x[0] = x_0
v = np.empty(N+1)
v[0] = v_0
```

```python
# Numerical integration cycle
for k in range(N) :
    x[k+1], v[k+1] = velocity_verlet(x[k], v[k], acc, dt)
```

```python
# Computation of kinetic, potential and total energy
# Particle's mass is taken as m=1 for simplicity
K = 0.5 * v**2 
P = 0.5 * (omega*x)**2
H = K+P

# Plotting energies
plt.plot(t, H, 'k-', label='total')
plt.plot(t, K, 'r-', label='kinetic')
plt.plot(t, P, 'b-', label='potential')
plt.xlabel('time')
plt.ylabel('energy')
plt.legend()
plt.show()
```

```python
# Comparison with analytical solution
x_exact = np.sin(t)
plt.plot(t, x, 'kx', label='numerical')
plt.plot(t, x_exact, 'b-', label='analytical')
plt.xlabel('time')
plt.ylabel('position')
plt.legend()
plt.show()
```

It seems VV is able to reproduce the analytical solution of our model problem. Total energy conservation is also obtained.

### Particle coupled to a system of independent harmonic oscillators

We now move on and apply VV to the following system:

$$ \begin{cases}
        m\ddot{x} + V'(x) + \sum_{k=1}^N\frac{C_k^2}{\omega_k^2}x = \sum_{k=1}^NC_kx_k \\
        \quad \\
        \ddot{x}_k+\omega_k^2x_k=C_kx & k=1,...,N
    \end{cases} \quad . $$

The oscillators' fequencies are discretized as:

$$ \omega_k = k\Delta\omega \; , \quad k=1,...,N \; , \quad \Delta\omega = \frac{\Omega}{N} \; , $$

while the coupling constants are obtained as:

$$ C_k = \sqrt{\frac{2\omega_k\Delta\omega J(\omega_k)}{\pi}} \quad . $$

In the following we will adopt an OOP approach; however VV could be implemented, in a less elegant way, with a procedural approach too.

```python
# Class for the ensemble of independent oscillators, plus the Brownian particle

class ensemble:
    
    def __init__(self, Omega, alpha, N, m, beta, v0):
        
        # Number of oscillators
        self.n_osc = N
        
        # High-frequency cutoff
        self.high_cut_off = Omega
        
        # Ohmic spectral density function
        self.spectral_dens = lambda w : w*alpha*(w<Omega)
        
        # Discrete nodal frequencies
        # The zero-th frequency would correspond to the one of the Browian particle and it's set to zero
        self.dw = Omega/N
        self.w = np.zeros(N+1)
        self.w[1:] = np.linspace(self.dw, Omega, N)
        
        # Coupling constants
        self.c = np.zeros(N+1)
        self.c[1:] = np.sqrt( 2.0*self.w[1:]*self.dw*self.spectral_dens(self.w[1:])/np.pi )
        # The zero-th coupling constant is the counter term (this simplify the implementation)
        self.c[0] = - np.sum( (self.c[1:]**2) / (self.w[1:]**2) )
        
        # Partial accelerations acting on a oscillator or on a Brownian particle
        self.mass = m
        self.acc_0 = lambda x : np.inner(self.c, x)/self.mass
        self.acc_k = lambda x, k : self.c[k]*x[0] - (self.w[k]**2)*x[k] 
        
        # Initial conditions
        self.x_init = np.zeros(N+1)
        self.v_init = np.zeros(N+1)
        
        # Initial velocities given initial kinetic energy
        K_s = 0.5*N/beta
        self.v_init = rng.normal(0.0, 1.0, N+1)
        self.v_init[1:] = np.sqrt( 2.0 * K_s / np.sum( self.v_init[1:]**2 ) ) * self.v_init[1:]
        self.v_init[0] = v0
        
        # Current state
        self.x_state = self.x_init
        self.v_state = self.v_init
        
        # Energies
        self.kin_ener = lambda : 0.5 * np.sum( self.v_state[1:]**2 ) + 0.5*self.mass*self.v_state[0]**2
        self.pot_ener = lambda : 0.5 * np.sum( ( self.w[1:]*self.x_state[1:] - \
                                                (self.c[1:]/self.w[1:])*self.x_state[0] )**2 )
        self.tot_ener = lambda : self.kin_ener() + self.pot_ener()
    
    # Total acceleration
    def acc(self, x) :
        a = np.zeros(self.n_osc+1)
        a[0] = self.acc_0(x)
        for k in range(1,self.n_osc+1) :
            a[k] = self.acc_k(x,k)
        return a
```

```python
# This class implements the numerical integrator

# It basically 'wraps' the VV scheme with some useful parameters (e.g. time step); it also computes energies
# on the fly

class integrator:
    
    def __init__(self, dt, T):
        self.time_step = dt
        self.time_max = T
        self.n_steps = int(T/dt)-1
        self.t_vec = np.linspace(0,T,self.n_steps+1)
        self.x_par = np.zeros(self.n_steps+1, dtype=float)
        self.v_par = np.zeros(self.n_steps+1, dtype=float)
        self.K = np.zeros(self.n_steps+1, dtype=float)
        self.P = np.zeros(self.n_steps+1, dtype=float)
        self.H = np.zeros(self.n_steps+1, dtype=float)
        
    def propagate(self, ens):
        self.x_par[0] = ens.x_init[0]
        self.v_par[0] = ens.v_init[0]
        self.K[0] = ens.kin_ener()
        self.P[0] = ens.pot_ener()
        self.H[0] = self.K[0] + self.P[0]
        for n in range(self.n_steps) :
            ens.x_state, ens.v_state = velocity_verlet(ens.x_state, ens.v_state, ens.acc, self.time_step)
            # Quantities to save
            self.x_par[n+1] = ens.x_state[0]
            self.v_par[n+1] = ens.v_state[0]
            self.K[n+1] = ens.kin_ener()
            self.P[n+1] = ens.pot_ener()
            self.H[n+1] = self.K[n+1] + self.P[n+1]
```

```python
# First test: vary the number of oscillators and observe what changes

# The OOP implementation allows to define different instances simultaneously (one for each test case).

int1 = integrator(dt=0.1, T=100.0)
ens1 = ensemble(Omega=1.0, alpha=1.0, N=10, m=0.1, beta=1.0, v0=1.0)
ens2 = ensemble(Omega=1.0, alpha=1.0, N=100, m=0.1, beta=1.0, v0=1.0)
ens3 = ensemble(Omega=1.0, alpha=1.0, N=1000, m=0.1, beta=1.0, v0=1.0)
```

```python
int1.propagate(ens1)
```
